```python
import instaloader
import pandas as pd
```

```python
USER = ''
PASSWORD = ''
```

```python
def get_comments_from_posts(SHORTCODE):
    try:
        L = instaloader.Instaloader()
        L.login(USER, PASSWORD)
        post = instaloader.Post.from_shortcode(L.context, SHORTCODE)

        comments = []
        get_comments = post.get_comments()
        for coment in get_comments:
            print(coment.owner.username)
            print(coment.text)
            print()
            new_comment = {
                'id_ig': coment.owner.userid,
                'username': coment.owner.username,
                'id_comment': coment.id,
                'fecha_comment': str(coment.created_at_utc),
                'comment': str(coment.text),
                'likes_comment': coment.likes_count
            }
            comments.append(new_comment)
            answers = coment.answers
            for a in answers:
                new_answer = {
                    'id_ig': a.owner.userid,
                    'username': a.owner.username,
                    'id_comment': a.id,
                    'fecha_comment': str(a.created_at_utc),
                    'comment': str(a.text),
                    'likes_comment': a.likes_count
                }
                comments.append(new_answer)

        return {
            "success": True,
            "msg": "method executed successfully",
            "body": comments,
        }
    except:
        return {
            "success": False,
            "msg": "exception created",
            "body": comments,
        }
```

```python
comments_ig_post = get_comments_from_posts(SHORTCODE='CReyMz0NNlz')
```

```python
df = pd.DataFrame(columns=['id_ig', 'username', 'id_comment', 'fecha_comment', 'text_comment', 'likes_comment'])
for comment in comments_ig_post['body']:
    id_ig = comment['id_ig']
    username = comment['username']
    id_comment = comment['id_comment']
    fecha_comment = comment['fecha_comment']
    text_comment = comment['comment']
    likes_comment = comment['likes_comment']
    
    df = df.append({
        'id_ig': id_ig,
        'username': username,
        'id_comment': id_comment,
        'fecha_comment': fecha_comment,
        'text_comment': text_comment,
        'likes_comment': likes_comment
    }, ignore_index=True)
```
